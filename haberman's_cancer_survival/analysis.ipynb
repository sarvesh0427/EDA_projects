## Univariant Analysis - Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
g = sns.FacetGrid(df, hue='survival_status',height=5)
g.map(sns.histplot, 'age', kde=True)
g.add_legend()
```

```python
sns.FacetGrid(df,hue='survival_status',height = 5)\
.map(sns.histplot, 'operation_year', kde = True)\
.add_legend()
```

```python
sns.FacetGrid(df,hue='survival_status',height = 5)\
.map(sns.histplot, 'aln', kde = True)\
.add_legend()
```

## Violin plot

```python
sns.violinplot(x='survival_status', y='age', data =df)
plt.show()
```

```python
sns.violinplot(x='survival_status', y='operation_year', data =df)
plt.show()
```

```python
sns.violinplot(x='survival_status', y='aln', data =df)
plt.show()
```

## Box plot

```python
sns.boxplot(x="survival_status", y='age', data =df)
plt.show()
```

```python
sns.boxplot(x="survival_status", y='operation_year', data =df)
plt.show()
```

```python
sns.boxplot(x="survival_status", y='aln', data =df)
plt.show()
```

## Pair plot

```python
sns.set_style("whitegrid")
sns.pairplot(df,hue= 'survival_status',height = 5)
plt.show()
```
